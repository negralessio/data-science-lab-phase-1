# Chi-Squared-Test for Categorical Features and Target Variable
We assume here that the target variable is categorical and therefore we can use the Chi Squared Test to test for independency between all categorical features and the target variable.

The Hypothesis are defined as follows ([Source, Accessed 22nd April](https://analyticsindiamag.com/a-beginners-guide-to-chi-square-test-in-python-from-scratch/))
- $H_0$: The two categorical variables have no relationship (independent)
- $H_1$: There is a relationship (dependent) between two categorical variables

The test statistic is given as ... $$ \chi^2 = \sum^m_{i=1} \sum^r_{j=1} \frac{(O_{ij} - E_{ij})^2}{E_{ij}}$$
where
- $O_{ij}$ is the observed value and
- $E_{ij}$ the expected value

We reject $H_0$ if $\chi^2 > \chi^2_{(1-\alpha), (m-1),(r-1)}$ to a significance level of $\alpha=0.05$

```python
# Define significance level
ALPHA = 0.05

# Create empty list to populate later
feat_1 = []
feat_2 = []
p_list = []
chi_list = []
dof_list = []
res_list = []

# Iterate through all categorical columns and perform chi square test
for feature in categoric_columns:
    # Create contingency table
    contingency_table = pd.crosstab(df_train_categoric[feature],
                                df_train_categoric["damage_grade"])
    # Perform chi square test and get test statistic, p-value, degrees of freedom and expected values
    chi, p, dof, expected = chi2_contingency(contingency_table)
    # Populate lists
    feat_1.append(feature)
    feat_2.append("damage_grade")
    p_list.append(p)
    chi_list.append(chi)
    dof_list.append(dof)
    res_list.append(p < ALPHA)

# Populate Result DataFrame and view it
chi_test_results = pd.DataFrame({"Feature 1": feat_1, "Feature 2": feat_2,
                                "Test Statistic" : chi_list, "Degrees of Freedom": dof_list,
                                "p-value": p_list, "Reject Nullhypothesis": res_list})
# Round to 4 decimal places
chi_test_results = chi_test_results.round(4)
chi_test_results
```

## Findings
- We can reject the null hypothesis $H_0$ for all categorical features with the target variable at a significance level of $\alpha = 0.05$ and therefore we have evidence that their are not independent.

# Mutual Information for Categorical Features and Target
Mutual Information is defined as ([Source, Accessed 22nd April](https://www.blog.trainindata.com/mutual-information-with-python/)): $$I(X, Y) = \sum_X \sum_Y p(x,y) \; log(\frac{p(x,y)}{p(x)p(y)}) $$
- where $p(x)$ is the marginal distribution and
- $p(x,y)$ the joint distribution

It measures the amount of information we can know from one variable by observing the values of the second variable.  In comparison to Pearson correlation, it it is able to measure any type of relationship between variables, not just linear associations.

```python
# Get categoric columns without target
categoric_columns_wo_target = [x for x in categoric_columns if x != "damage_grade"]

# Empty list to populate later
mi_score: list = []

# Iterate through all categorical columns and perform chi square test
for feature in categoric_columns_wo_target:
    # Create contingency table
    contingency_table = pd.crosstab(df_train_categoric[feature], df_train_categoric["damage_grade"])
    # Compute mutual information score between each categorical feature and target
    score = mutual_info_score(labels_true=None, labels_pred=None, contingency = contingency_table)
    mi_score.append(score)

mi_df_results = pd.DataFrame({"Feature": categoric_columns, "Mutual Information": mi_score})
```

## Visualize Mutual Information Results

```python
plt.figure(figsize=(12, 5), dpi=256)
# Compute mean and standard deviation
avg_mi = round(np.mean(mi_df_results["Mutual Information"].values), 4)
std_mi = round(np.std(mi_df_results["Mutual Information"].values), 4)
plt.title(f"Mutual Information between Categorical Feature and Target ($\mu$={avg_mi}, $\sigma$={std_mi})", size=16)
g = sns.barplot(data=mi_df_results, x="Feature", y="Mutual Information")
g.bar_label(g.containers[0], padding=1.5)
plt.axhline(y=avg_mi, c="red", linestyle="--")
plt.xticks(rotation=45)
plt.show()
```

## Findings
- Foundation Type has the highest Mutual Information to the target, i.e. there is a strong relationship between the type of foundation and the damage grade
- Analog for ground_floor_type, i.e. the type of the ground floor
- Analog for other_floor_type, i.e. the type of constructions used in higher than the ground floors (except of roof)
- Analog for roof_type
